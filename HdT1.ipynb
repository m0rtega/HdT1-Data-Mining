```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
db = pd.read_csv('tmdb-movies.csv')
db.head()
```

```python
x = db['budget']
y = db['revenue']

corr = plt
corr.scatter(x, y, alpha=0.5)
corr.gcf().set_size_inches(12, 9)
corr.title('CorrelaciÃ³n Budget & Revenue', size=24)
corr.xlabel('Budget (cientos de millones)', size=18)
corr.yscale("symlog")
corr.ylabel('Revenue (billones)', size=18)
corr.show()
```

```python
dates = pd.to_datetime(db['release_date'])
x = dates.dt.month
y = db['revenue']

dateGraph = plt
db.revenue.groupby(dates.dt.month).mean().plot(kind="bar")
dateGraph.gcf().set_size_inches(12, 9)
dateGraph.title('Revenue per month', size=24)
dateGraph.xlabel('Months', size=18)
dateGraph.ylabel('Revenue (decenas de millones)', size=18)
dateGraph.show()
```

```python
db_genres = db.copy()
genres = db_genres['genres'].str.split('|').apply(pd.Series, 1).stack().reset_index(level=1, drop=True)
genres.name = 'genre_split'
db_genres = db_genres.drop(['genres'], axis=1).join(genres)
```

```python
db_genres.head(5)
```

```python
sum(db_genres.duplicated())
```

```python
db_genres.runtime.groupby(db_genres['genre_split']).mean().plot(kind='bar')
plt.gcf().set_size_inches(12, 9)
plt.title('Runtime per genre', size=24)
plt.xlabel('Genres', size=18)
plt.ylabel('Runtime (minutes)', size=18)
plt.show()
```
